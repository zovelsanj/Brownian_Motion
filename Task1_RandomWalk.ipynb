## RANDOM WALK

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
MODE = ["1D", "2D", "3D"]
N = 1000
```

```python
def random_walk(MODE):
    def walk():
        steps[steps<=0.5] = -1
        steps[steps>0.5] = 1
        
    if MODE == "1D":
        steps = np.random.random(N)
        # type(steps)
        for i in range(len(steps)): walk()
        plt.title(f'{MODE} RANDOM WALK')
        plt.plot(np.cumsum(steps))

    if MODE == "2D":
        steps = np.random.random((N,2))
        walk()
        x = np.cumsum(steps, axis=0)[:,0]
        y = np.cumsum(steps, axis=0)[:,1]
        plt.title(f'{MODE} RANDOM WALK')
        plt.plot(x,y, linestyle = '', marker = '.')
        
    if MODE == "3D":
        steps = np.random.random((N,3))
        walk()
        x = np.cumsum(steps, axis=0)[:,0]
        y = np.cumsum(steps, axis=0)[:,1]
        z = np.cumsum(steps, axis=0)[:,2]
        fig = plt.figure(figsize=(20,20))
        fig.add_subplot(121, projection='3d')
        ax = plt.gca()
        ax.scatter(x, y, z, marker = 'o')        
        
```

```python
random_walk("1D")
```

```python
random_walk("2D")
```

```python
random_walk("3D")
```

## TRAJECTORY

```python

def trajectory():
    steps = np.random.random((N, N))
    steps[steps<=0.5] = -1
    steps[steps>0.5] = 1
    cumulative = np.cumsum(steps, axis=0)
    plt.plot(range(N), cumulative)
    plt.plot(range(N), np.average(cumulative, axis=1),color='black')
    plt.title("RANDOM WALK TRAJECTORIES")
```

```python
trajectory()
```
