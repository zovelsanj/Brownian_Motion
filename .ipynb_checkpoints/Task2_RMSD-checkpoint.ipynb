## RMSD OF TRAJECTORIES

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 700
N_traj = 500
RMS = np.zeros(N_traj)
# suppose inital position be origin
def rms_distance(a):   
    steps = np.random.random((N_traj, N))
    steps[steps<=0.5] = -a
    steps[steps>0.5] = a
    mean = np.average(np.cumsum(steps))
#     plt.plot(range(N_traj), np.cumsum(steps, axis=0))
    return mean**2
```

```python
rms_distance(1)
plt.plot(range(N_traj), rms_distance(1))
```

```python
rms = rms_distance(1)
slope, intercept = np.polyfit(range(N_traj), rms, 1)
y = [slope*i+intercept for i in range(N_traj)]
plt.plot(range(N_traj), y)
plt.plot(range(N_traj),rms)
plt.title("Fit data")
```
