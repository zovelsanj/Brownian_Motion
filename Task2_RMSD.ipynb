## RMSD OF TRAJECTORIES

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 1000
N_traj = 500
msd = np.zeros(N_traj)
# suppose inital position be origin
def msd(a):   
    steps = np.random.random((N_traj, N))
    steps[steps<=0.5] = -a
    steps[steps>0.5] = a
    MSD = np.average(np.cumsum(steps, axis=1)**2, axis=0)
    return MSD
```

```python
# msd(1)
plt.plot(range(N), msd(1))
plt.title("Mean Square Distance")
plt.xlabel("Number of Steps")
plt.ylabel("MSD")
```

```python
rms = np.sqrt([msd(i) for i in range(1,6)])
slope, intercept = np.polyfit(range(N), rms[0], 1)
y = [slope*i+intercept for i in range(N)]
plt.plot(range(N), y)
plt.plot(range(N),rms[0])
plt.legend(["Fit distance","RMS distance" ])
plt.title("RMS Fit data")
```

```python
for k in range(1,5):
    slope, intercept = np.polyfit(range(N), rms[k], 1)
    y = [slope*i+intercept for i in range(N)]
    plt.plot(range(N), y)
    plt.legend([1,2,3,4,5])
```

```python
import random
def brownianMotion(timePoints):
    Trajectory = [0]
    for t in range(len(timePoints)-1):
        randomNumber = random.gauss(0, timePoints[t+1]-timePoints[t])
        Trajectory.append(Trajectory[t] + randomNumber)
    return Trajectory
```
